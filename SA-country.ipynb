```python
import requests
wiki_link="https://en.wikipedia.org/wiki/List_of_South_American_countries_by_area"
link=requests.get(wiki_link).text
```

```python
print(link)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(link, 'lxml')
print(soup)
```

```python
print(soup.prettify())
```

```python
soup.title
```

```python
soup.title.string
```

```python
soup.a
```

```python
soup.find_all("a")
```

```python
alla_link=soup.find_all("a")
for link in alla_link:
    print(link.get("href"))
```

```python
all_tables=soup.find_all('table')
print(all_tables)
```

```python
right_table=soup.find('table', class_='wikitable sortable')
right_table
```

```python
table_links = right_table.findAll('a')
table_links
```

```python
country = []
for links in table_links:
    country.append(links.get('title'))
print(country)
```

```python
import pandas as pd
df=pd.DataFrame()
df['SAcountry']=country
df
```
